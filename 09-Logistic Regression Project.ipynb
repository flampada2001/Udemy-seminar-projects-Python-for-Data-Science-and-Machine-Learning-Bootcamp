# Logistic Regression Project 

In this project we will be working with a fake advertising data set, indicating whether or not a particular internet user clicked on an Advertisement. We will try to create a model that will predict whether or not they will click on an ad based off the features of that user.

This data set contains the following features:

* 'Daily Time Spent on Site': consumer time on site in minutes
* 'Age': cutomer age in years
* 'Area Income': Avg. Income of geographical area of consumer
* 'Daily Internet Usage': Avg. minutes a day consumer is on the internet
* 'Ad Topic Line': Headline of the advertisement
* 'City': City of consumer
* 'Male': Whether or not consumer was male
* 'Country': Country of consumer
* 'Timestamp': Time at which consumer clicked on Ad or closed window
* 'Clicked on Ad': 0 or 1 indicated clicking on Ad

# Logistic Regression

Now it's time to do a train test split, and train our model!

You'll have the freedom here to choose columns that you want to train on!

**Split the data into training set and testing set using train_test_split**

```python
from sklearn.model_selection import train_test_split
```

```python
ad_data.info()
```

```python
sns.heatmap(ad_data.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

__'First of all we see that there are no NaN values!__

__1)__'Ad topic Line', 'City', 'Country' and 'Timestamp' are all string like so we remove them!

```python
ad_data.drop(['Ad Topic Line','City', 'Country','Timestamp'],axis=1, inplace=True)
```

```python
ad_data.head()
```

__2)__ 'Male' and 'Clicked on Ad' are 0 and 1 we turn them to True and False! 

We then replace the new columns with the old ones!

```python
Male=pd.get_dummies(ad_data['Male'],drop_first=True,prefix='Male')
Male.columns=['Male'] #Renaming the column because it just sets it to 1!

Clicked_on_Ad=pd.get_dummies(ad_data['Clicked on Ad'],drop_first=True,prefix='Clicked on Ad') 
Clicked_on_Ad.columns=['Clicked on Ad'] #Renaming the column because it just sets it to 1!

ad_data.drop(['Male','Clicked on Ad'],axis=1,inplace=True)
ad_data=pd.concat([ad_data,Male,Clicked_on_Ad],axis=1) # REMEMBER TO PUT THE ORIGINAL DF IN CONCAT!!
```

```python
ad_data.head()
```

### PS: There is no need to turn the values in true or False, LogisticRegressin can handle 0s and 1s just fine!

Our __y__ is the __'Clicked on Ad'__ column!

```python
X_train, X_test, y_train, y_test = train_test_split(ad_data.drop('Clicked on Ad',axis=1),ad_data['Clicked on Ad'],test_size=0.30,random_state=101)
```

**Train and fit a logistic regression model on the training set.**

```python
from sklearn.linear_model import LogisticRegression
```

```python
ad_model=LogisticRegression()
ad_model.fit(X_train,y_train)
```

## Predictions and Evaluations
**Now predict values for the testing data.**

```python
ad_pred=ad_model.predict(X_test)
ad_pred
```

**Create a classification report for the model.**

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,ad_pred))
```

```python
confusion_matrix(y_test,ad_pred)
```
